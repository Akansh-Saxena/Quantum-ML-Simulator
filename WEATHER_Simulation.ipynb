### Step 2: Load the Dataset
Since actual subatomic antigravity data is restricted (or theoretical), we simulate a high-fidelity dataset of 10,000 quantum experiments.

**Physics Context:**
- **Electromagnetic Field Strength (Tesla)**: High-intensity magnetic fields used to stabilize the exotic matter.
- **Exotic Matter Density (kg/m^3)**: The concentration of negative-mass matter or metamaterials.
- **Subatomic Resonance (THz)**: The frequency at which the material's atomic lattice vibrates.
- **Energy Input (Terawatts)**: The raw power injected into the containment field.

**Target Variable**:
- **Gravitational Disruption Index (GDI)**: A dimensionless scalar where 0 represents normal Earth gravity (1g) and negative values represent localized gravitational anomalies.

We use a non-linear physics-based formula combined with quantum noise to generate the GDI:
$$GDI = 0.05 \cdot (\text{EM\_Field} \cdot \ln(\text{Density} + 1)) + 0.001 \cdot (\text{Resonance}^2 \cdot \text{Energy}) + \text{Noise}$$

```python
# Step 2: Generate Synthetic Dataset
np.random.seed(42)
num_samples = 10000

# Features
em_field_strength = np.random.uniform(50, 500, num_samples) # in Tesla
exotic_matter_density = np.random.uniform(0.01, 10.0, num_samples) # kg/m^3
subatomic_resonance = np.random.uniform(1, 100, num_samples) # THz
energy_input = np.random.uniform(0.5, 50.0, num_samples) # Terawatts

# Theoretical Non-linear Physics Equation for GDI
quantum_noise = np.random.normal(0, 2.5, num_samples)
gdi = 0.05 * (em_field_strength * np.log(exotic_matter_density + 1)) + \
      0.001 * (np.power(subatomic_resonance, 2) * energy_input) + \
      quantum_noise

df = pd.DataFrame({
    'EM_Field_Strength_Tesla': em_field_strength,
    'Exotic_Matter_Density': exotic_matter_density,
    'Subatomic_Resonance_THz': subatomic_resonance,
    'Energy_Input_Terawatts': energy_input,
    'Gravitational_Disruption_Index': gdi
})

print("Dataset generated successfully. Shape:", df.shape)
display(df.head())
```

### Step 3: Check Missing Values
Sensor anomalies and temporary containment breaches often cause missing readouts in quantum experiments. We simulate this by introducing random `NaN` values into our dataset.

```python
# Step 3: Introduce simulated data corruption
np.random.seed(42)

# Corrupt roughly 5% of the features randomly
for col in df.columns[:-1]: # Don't corrupt the target variable (GDI)
    mask = np.random.rand(num_samples) < 0.05
    df.loc[mask, col] = np.nan

print("Missing values after simulating sensor anomalies:")
print(df.isnull().sum())
```

### Step 4: Data Preprocessing
We must handle the missing values so our neural network doesn't collapse. We will impute the missing values with the median of each respective column to maintain distribution integrity.
After imputation, we apply `StandardScaler` to ensure all features have a mean of 0 and a variance of 1. This prevents high-magnitude variables (like Energy Input) from dominating the gradients during Neural Network training.

```python
# Step 4: Handle missing values via Impute with median
df_clean = df.fillna(df.median())

print("Missing values after imputation:")
print(df_clean.isnull().sum())

# Separate features and target
X = df_clean.drop('Gravitational_Disruption_Index', axis=1)
y = df_clean['Gravitational_Disruption_Index']

# Advanced feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for better visibility
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
display(X_scaled_df.head())
```

### Step 5: Split the Data
We split the data into 80% training and 20% testing sets. The training set is used to optimize the model weights, while the testing set verifies that our model generalizes well to unseen quantum states.

```python
# Step 5: 80/20 train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

### Step 6: Build and Train the Model
We will implement a futuristic **Multi-Layer Perceptron (MLP) Deep Neural Network** to capture the highly non-linear dynamics of antigravity variables.
The network architecture consists of:
- An input layer mapping the 4 quantum features.
- Three dense hidden layers with ReLU activation functions, decreasing in complexity to extract hierarchical non-linear patterns.
- A final output layer with a linear activation function for regression (predicting the exact GDI value).
- Adam optimizer and Mean Squared Error (MSE) loss function.

```python
# Step 6: Build the futuristic MLP Neural Network
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=[X_train.shape[1]]),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1) # Linear output for regression
])

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

model.summary()

# Train the model
print("\n--- Initiating Neural Network Training Sequence ---")
history = model.fit(
    X_train, y_train,
    epochs=50,
    validation_split=0.2,
    batch_size=32,
    verbose=1
)
print("--- Training Sequence Complete ---")
```

### Step 7: Evaluate the Model
We will evaluate the trained MLP network using the untouched test dataset.
Key metrics:
- **Mean Squared Error (MSE)**: The average squared difference between estimated and actual GDI.
- **Root Mean Squared Error (RMSE)**: Easier to interpret as it's in the same unit as the GDI.
- **Mean Absolute Error (MAE)**: The average absolute error.
- **R-Squared (R²)**: The proportion of the variance in GDI predictable from the features. A value close to 1.0 indicates near-perfect prediction capabilities.

```python
# Step 7: Model Evaluation
y_pred = model.predict(X_test).flatten()

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\n--- Quantum Simulation Model Diagnostics ---")
print(f"Mean Squared Error (MSE):       {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE):      {mae:.4f}")
print(f"R-Squared (R2):                 {r2:.4f}")
print("--------------------------------------------")
```

### Step 8: Visualize the Results
To properly demonstrate the highly complex, non-linear relationship between our input parameters, we generate a futuristic, interactive 3D surface plot.
This map charts how **Electromagnetic Field Strength** and **Energy Input** interact to modulate the **Gravitational Disruption Index**, holding other variables at their mean values.

*Note: You can rotate, zoom, and explore this 3D map to find the "sweet spot" for maximum antigravity generation.*

```python
# Step 8: Interactive 3D Surface Visualization using Plotly
# Create a grid for EM Field and Energy Input
em_field_range = np.linspace(X['EM_Field_Strength_Tesla'].min(), X['EM_Field_Strength_Tesla'].max(), 50)
energy_range = np.linspace(X['Energy_Input_Terawatts'].min(), X['Energy_Input_Terawatts'].max(), 50)

em_grid, energy_grid = np.meshgrid(em_field_range, energy_range)

# Hold Density and Resonance at their median values for the simulation surface
density_median = df_clean['Exotic_Matter_Density'].median()
resonance_median = df_clean['Subatomic_Resonance_THz'].median()

# Flatten grids to pass into the model
em_flat = em_grid.flatten()
energy_flat = energy_grid.flatten()
density_flat = np.full_like(em_flat, density_median)
resonance_flat = np.full_like(em_flat, resonance_median)

# Create an artificial test DataFrame with the exact column order
sim_df = pd.DataFrame({
    'EM_Field_Strength_Tesla': em_flat,
    'Exotic_Matter_Density': density_flat,
    'Subatomic_Resonance_THz': resonance_flat,
    'Energy_Input_Terawatts': energy_flat
})

# Scale features based on the previously fit scaler
sim_scaled = scaler.transform(sim_df)

# Predict GDI using the Neural Network
gdi_pred = model.predict(sim_scaled).flatten()
gdi_surface = gdi_pred.reshape(em_grid.shape)

# Create futuristic 3D Plot
fig = go.Figure(data=[go.Surface(
    z=gdi_surface, 
    x=em_grid, 
    y=energy_grid, 
    colorscale='Plasma',
    lighting=dict(ambient=0.8, diffuse=0.8, roughness=0.1, specular=1.0, fresnel=0.8)
)])

fig.update_layout(
    title='Quantum Simulation: Gravitational Disruption Manifold',
    scene=dict(
        xaxis_title='EM Field Strength (Tesla)',
        yaxis_title='Energy Input (TW)',
        zaxis_title='Predicted GDI',
        xaxis=dict(gridcolor='cyan', zerolinecolor='cyan', backgroundcolor='black'),
        yaxis=dict(gridcolor='magenta', zerolinecolor='magenta', backgroundcolor='black'),
        zaxis=dict(gridcolor='yellow', zerolinecolor='yellow', backgroundcolor='black')
    ),
    paper_bgcolor='black',
    font=dict(color='white'),
    margin=dict(l=0, r=0, b=0, t=50)
)

fig.show()
```
